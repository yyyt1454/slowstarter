```python
def get_preprocessed_df(df):
    df_copy = df.copy()
    stays = list(df_copy.columns)
    for stay in stays:
        df_copy[stay] = df_copy[stay].apply(lambda x: 1 if x>0 else 0)
    return df_copy
```

# what is mathing ID?

매칭 아이디는 기본적으로 하나의 여행에 한 개씩 부여됨


그러나 한 여행에 여러 도를 다닌 경우(여행한 광역시/도(해외여행: 국가) 컬럼의 코드가 다른 경우)에는 
다른 row로 분리되어 동일한 가구원이 여러 개의 매칭 아이디를 가지기도 함

15년도 데이터에는 매칭 아이디가 따로 부여되지 않았으며 '여행 월'과 '여행 번째'로 단위 여행이 구분됨

그러나 16년도, 17년도 데이터와 병합하는 과정에서 '여행 번째'를 '매칭 ID'로 바꾸었음. (컬럼명 통일을 위해서)

따라서 단위여행을 구분하기 위해서는 '매칭 ID'와 '여행 월'을 같이 봐야 함
