```python
# 컬럼 네임 재설정 

강원 = 강원.rename(columns = {'(2014년 ~ 2018년)':'2014년', 
                     'Unnamed: 8':'2015년', 
                     'Unnamed: 9':'2016년',
                     'Unnamed: 10':'2017년',
                     'Unnamed: 11':'2018년'})
강원.drop(0, inplace=True)

충북 = 충북.rename(columns = {'(2014년 ~ 2018년)':'2014년', 
                     'Unnamed: 8':'2015년', 
                     'Unnamed: 9':'2016년',
                     'Unnamed: 10':'2017년',
                     'Unnamed: 11':'2018년'})
충북.drop(0, inplace=True)

충남 = 충남.rename(columns = {'(2014년 ~ 2018년)':'2014년', 
                     'Unnamed: 8':'2015년', 
                     'Unnamed: 9':'2016년',
                     'Unnamed: 10':'2017년',
                     'Unnamed: 11':'2018년'})
충남.drop(0, inplace=True)

전북 = 전북.rename(columns = {'(2014년 ~ 2018년)':'2014년', 
                     'Unnamed: 8':'2015년', 
                     'Unnamed: 9':'2016년',
                     'Unnamed: 10':'2017년',
                     'Unnamed: 11':'2018년'})
전북.drop(0, inplace=True)

전남 = 전남.rename(columns = {'(2014년 ~ 2018년)':'2014년', 
                     'Unnamed: 8':'2015년', 
                     'Unnamed: 9':'2016년',
                     'Unnamed: 10':'2017년',
                     'Unnamed: 11':'2018년'})
전남.drop(0, inplace=True)

경북 = 경북.rename(columns = {'(2014년 ~ 2018년)':'2014년', 
                     'Unnamed: 8':'2015년', 
                     'Unnamed: 9':'2016년',
                     'Unnamed: 10':'2017년',
                     'Unnamed: 11':'2018년'})
경북.drop(0, inplace=True)

경남 = 경남.rename(columns = {'(2014년 ~ 2018년)':'2014년', 
                     'Unnamed: 8':'2015년', 
                     'Unnamed: 9':'2016년',
                     'Unnamed: 10':'2017년',
                     'Unnamed: 11':'2018년'})
경남.drop(0, inplace=True)

제주 = 제주.rename(columns = {'(2014년 ~ 2018년)':'2014년', 
                     'Unnamed: 8':'2015년', 
                     'Unnamed: 9':'2016년',
                     'Unnamed: 10':'2017년',
                     'Unnamed: 11':'2018년'})
제주.drop(0, inplace=True)
```

```python
# 통합데이터 생성

total = pd.concat([강원, 충북, 충남, 전북, 전남, 경북, 경남, 제주], axis = 0, ignore_index = True)
```

```python
# 결측치 0으로 처리

for i in range(0,12):
    for j in range(0, len(total)):
        if total.iloc[j][i] == '-':
            total.iloc[j][i] = 0
```

```python
# 방문객 수가 string으로 처리되어 있으므로 float 형식으로 변환

for i in range(6, 12):
    for j in range(0, len(total)):
        if type(total.iloc[j][i]) == str:
            total.iloc[j][i] = total.iloc[j][i].replace(',', '')
            total.iloc[j][i] = float(total.iloc[j][i])
        elif type(total.iloc[j][i]) == int:
            total.iloc[j][i] = float(total.iloc[j][i])
```

```python
for i in range(6, 11):
    for j in range(1, len(total)):
        print(type(total.iloc[j][i]))
```

```python
# 내국인 수치만 추출
total = total.loc[total['내/외국인'] == '내국인', :]
```

```python
# 2014년 ,2018년, 총계(5년치 합) 컬럼 삭제 

total = total.drop(total[['2014년', '2018년', '총계']], axis = 1)
```

```python
# 관광객 수가 집계되지 않은 컬럼?

total.loc[(total['2015년']==0) |(total['2016년']==0)|(total['2017년']==0), : ]
```

```python
# 집계 안된 년도가 뒤죽박죽 -> 일관된 처리 어려우므로 모두 삭제

total = total.loc[(total['2015년']!=0) & (total['2016년']!=0) & (total['2017년']!=0), : ]
```

```python
# 시군별 방문객 수 합계 테이블
totalsum = total.pivot_table(index = ['시도', '군구'], aggfunc = 'sum',
                values = ['2015년','2016년','2017년'])
```

```python
# 지역 별 관광지 수를 나타내는 컬럼 생성
count = pd.DataFrame(pd.value_counts(total['군구']))
count.reset_index(inplace = True)
count.columns = ['군구', '관광지수']
totalsum = pd.merge(totalsum, count, on = '군구')
```

```python
# 3년 간 관광객 수 총합 / 관광지 수 (관광지가 많은 도시)

totalsum['평균관광객수'] = (totalsum['2015년']+totalsum['2016년']+totalsum['2017년'])/totalsum['관광지수']
```

```python
# 비수도권지역 비인기 관광지 추출


비인기관광지 = totalsum.loc[totalsum['평균관광객수'] < totalsum['평균관광객수'].mean(), :].sort_values(by = '평균관광객수')

print(비인기관광지)
print(len(비인기관광지))
```
